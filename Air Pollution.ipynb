```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Training Data/Train.csv')
dft=pd.read_csv('Testing Data/Test.csv')
```

```python
def hypothesis(x, i, theta):
    return theta[0]+theta[1]*x['feature_1'][i]+theta[2]*x['feature_2'][i]+theta[3]*x['feature_3'][i]+theta[4]*x['feature_4'][i]+theta[5]*x['feature_5'][i]

def error(x, theta):
    m=x.shape[0]
    e=0
    for i in range(m):
        hx=hypothesis(x, i, theta)
        e+=(hx-x['target'][i])**2
    return e

def gradient(x, theta):
    g=np.zeros((6,))
    m=x.shape[0]
    for i in range(m):
        hx=hypothesis(x, i, theta)
        g[0]+=hx-x['target'][i]
        g[1]+=(hx-x['target'][i])*x['feature_1'][i]
        g[2]+=(hx-x['target'][i])*x['feature_2'][i]
        g[3]+=(hx-x['target'][i])*x['feature_3'][i]
        g[4]+=(hx-x['target'][i])*x['feature_4'][i]
        g[5]+=(hx-x['target'][i])*x['feature_5'][i]
    return g

def gradient_dec(x, lr=0.0001):
    theta=np.array([-2.0, 0.0, 1.0, -1.0, 2.0, 3.0])
    itr=0
    maxitr=100
    error_list=[]
    theta_list=[]
    while(itr<=maxitr):
        g=gradient(x, theta)
        e=error(x, theta)
        error_list.append(e)
        
        theta_list.append((theta[0], theta[1], theta[2], theta[3], theta[4], theta[5]))
        theta[0]-=lr*g[0]
        theta[1]-=lr*g[1]
        theta[2]-=lr*g[2]
        theta[3]-=lr*g[3]
        theta[4]-=lr*g[4]
        theta[5]-=lr*g[5]
        itr+=1
    return theta, error_list, theta_list
```

```python
theta, error_l, theta_l=gradient_dec(df)
```

```python
plt.plot(error_l)
plt.show()
```
